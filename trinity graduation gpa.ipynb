```python
# tkinter를 사용하기 위한 import
from tkinter import *
from tkinter import ttk
from selenium import webdriver
import time

# tkinter 객체 생성
window = Tk()

# 사용자 id와 password를 저장하는 변수 생성
user_id, password = StringVar(), StringVar()

# 사용자 id와 password를 비교하는 함수
def go_trinity():
    driver=webdriver.Chrome('C:\Chrome_Driver\chromedriver.exe')

    # 크롬을 통해 트리니티 로그인 화면에 접속
    driver.get('https://uportal.catholic.ac.kr/sso/jsp/sso/ip/login_form.jsp')

    # 아이디와 비밀번호를 입력 (0.5초 씩 기다린 후 실행)
    time.sleep(0.5)
    driver.find_element_by_name('userId').send_keys(user_id.get())
    time.sleep(0.5)
    driver.find_element_by_name('password').send_keys(password.get())

    # Xpath를 이용하여 로그인을 시도
    driver.find_element_by_xpath('/html/body/div/form/div/div/div[1]/dl/dd[3]/button').click()

    time.sleep(3)
    from bs4 import BeautifulSoup

    driver.get('https://uportal.catholic.ac.kr/stw/main/noti/mainNoti.do')
    #table=driver.find_element_by_name('myinfo-table')
    #tbody=table.find_element_by_tag_name('tbody')
    #rows=tbody.find_elements_by_tag_name('tr')
    #for index, value in enumerate(rows):
    #    body=value.find_elements_by_tag_name('td')[0]
    #    print(body.text)


    ## 자기관리 역량 과목
    time.sleep(3)
    driver.find_element_by_xpath('//*[@id="JsrMain3c"]/main/div/div[1]/section/div/div[2]/table/tbody/tr[2]/td').click()

    html=driver.page_source
    soup=BeautifulSoup(html,'lxml')

    self=soup.find_all('div','ucups-grid-row')

    for subjects in self:
        print(subjects.ucups-title)
        
    return

# id와 password, 그리고 확인 버튼의 UI를 만드는 부분
ttk.Label(window, text = "Username : ").grid(row = 0, column = 0, padx = 10, pady = 10)
ttk.Label(window, text = "Password : ").grid(row = 1, column = 0, padx = 10, pady = 10)
ttk.Entry(window, textvariable = user_id).grid(row = 0, column = 1, padx = 10, pady = 10)
ttk.Entry(window, textvariable = password,show="*").grid(row = 1, column = 1, padx = 10, pady = 10)
ttk.Button(window, text = "Login", command = go_trinity).grid(row = 2, column = 1, padx = 10, pady = 10)

window.mainloop()
```
